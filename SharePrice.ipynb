# Downloading data and saving to csv

```python

tickers = {}
history_data = {}

for share in shares_list:
    tickers[share] = yf.Ticker(share)
    history = tickers[share].history(period="max")
    history_data[share] = history

    file_path = os.path.join("/notebooks/data_csv", f"{share}_data.csv")
    history.to_csv(file_path)
```

# Preprocessing data
Adding rolling horizons, ratio columns, tomorrow close price, deleting dividens and stock splits

```python
import pandas as pd
import os

horizons = [2, 5, 60, 250, 1000]
new_predictors = []
large_shares = []
all_data = []

for share in shares_list:

    file_path = os.path.join("/notebooks/data_csv", f"{share}_data.csv")
    history_share = pd.read_csv(file_path, index_col=0)

    history_share['Share_Name'] = share

    if '1990-01-01' in history_share.index:
        print(f'Data for {share} does contain a 1990-01-01')
    else:
        print(f'Data for {share} does not contain a 1990-01-01')

    if "Dividends" in history_share.columns and "Stock Splits" in history_share.columns:
        del history_share["Dividends"]
        del history_share["Stock Splits"]
    else:
        continue
    history_share['Tomorrow'] = history_share['Close'].shift(-1)
    history_share['Target'] = (history_share['Tomorrow'] > history_share['Close']).astype(int)

    for horizon in horizons:

        rolling_averages_close = history_share["Close"].rolling(horizon).mean()

        ratio_column = f"Close_Ratio_{horizon}"
        history_share[ratio_column] = history_share["Close"] / rolling_averages_close
        trend_column = f"Trend_{horizon}"
        history_share[trend_column] = (history_share["Target"].shift(1).rolling(horizon).sum())

    history_share = history_share.dropna().copy()
    
    if len(history_share) > 3000:
        large_shares.append(share)
        all_data.append(history_share)
        
    history_share.to_csv(file_path)

if all_data:
    all_data_df = pd.concat(all_data)
    file_path = os.path.join("/notebooks/data_csv", f"all_data_df.csv")
    all_data_df.to_csv(file_path)
else:
    print("No dataframes with more than 3000 rows found.")

print("Shares with more than 3000 rows: ", large_shares)
```

# Train model
Function to train model with grid search, 

```python
def train_model_with_gridsearch(model, param_grid, data, predictors, test_size=0.2):
    
    best_model = None
    for name, group in data.groupby('Share_Name'):
        print(f"Processing share: {name}")
        
        if isinstance(model, CatBoostRegressor):
            cat_features_indices = np.where(group[predictors].dtypes != float)[0]
            fit_params = {'cat_features': cat_features_indices}
            
        elif isinstance(model, XGBRegressor) or isinstance(model, LGBMRegressor):
            le = LabelEncoder()
            group['Share_Name'] = le.fit_transform(group['Share_Name'])
            fit_params = {}
            
        else:
            raise ValueError("The model must be either CatBoostRegressor or XGBRegressor or LGBMRegressor")

        split_index = int(group.shape[0] * (1-test_size))
        train = group.iloc[:split_index]
        test = group.iloc[split_index:]

        grid_search = GridSearchCV(model, param_grid, cv=5, error_score='raise')
        grid_search.fit(train[predictors], train["Tomorrow"], **fit_params)

        best_model = grid_search.best_estimator_

        predictions = best_model.predict(test[predictors])

        error_rate = mape(test["Tomorrow"], predictions)

        print(f"Best parameters: {grid_search.best_params_}")
        print(f"Mean Absolute Percentage Error (MAPE): {error_rate}%")
    return best_model
```

```python
model = xgb.XGBRegressor(tree_method='gpu_hist', verbosity=2)


for name, group in all_data_df.groupby('Share_Name'):
    print(f"Processing share: {name}")
    start = time.time()
    
    best_model = train_model_with_gridsearch(model, xgb_param_grid, group, predictors)
    
    end = time.time()
    elapsed_time = end - start
    print(f"The task took {elapsed_time} seconds to complete.")
model_path = os.path.join('/notebooks/models', 'final_model.pkl')
joblib.dump(best_model, model_path)
```
